```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("Features-data-set.csv")
data1['Date'] = pd.to_datetime(data1['Date'])
data1.sort_values(["Date" , "Store"])
```

```python
data2 = pd.read_csv("sales-data-set.csv")
data2['Date'] = pd.to_datetime(data2['Date'])
data2.sort_values(["Date" , "Store"])
```

```python
data3 = pd.read_csv("stores-data-set.csv")
```

```python
data = pd.merge(data1, data2[["Date" , "Store" , "Weekly_Sales" , "Dept"]] ,on=['Date', 'Store'])
data = pd.merge(data , data3 , on = "Store")
data.sort_values(["Date" , "Store"])
data.sort_values(["Date" , "Store"])
```

```python
my_list = []
for i in data.index:
    if data.loc[i , "Type"] not in my_list:
        my_list.append(data.loc[i , "Type"])
q , w = pd.factorize(data["Type"])
data["Type"] = q
```

```python
x = data[["Store", "Dept","Temperature" ,"MarkDown1","MarkDown2","MarkDown3","MarkDown4","MarkDown5", "Fuel_Price" , "CPI" , "Unemployment" , "IsHoliday" , "Type" , "Size"]]
y = data["Weekly_Sales"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state =123)
```

```python
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(random_state=0)
regr.fit(x_train , y_train)
y_hat = regr.predict(x_test)

from sklearn.metrics import r2_score
print("r2_score:" , "\n" ,r2_score(y_test , y_hat) , "\n")
print("Training Score:\n",regr.score(x_train,y_train)*100)
```
